# SuperMart Grocery Sales Analysis


# Importing all necessary libraries and uploading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Supermart Grocery Sales - Retail Analytics Dataset.csv')
df
```

```python
df["Year"] = pd.to_datetime(df['Order Date']).dt.year
df["Day"] = pd.to_datetime(df['Order Date']).dt.day_name()
df["Month"] = pd.to_datetime(df['Order Date']).dt.month_name()
```

# Visualisation


```python
df.groupby("Year").sum().plot(kind = "bar")
```

```python
df.groupby("Year").sum()["Sales"].plot(kind = "bar")
```

```python
np.unique(df["Category"])
```

```python
semi_d = df[df["Category"]== "Bakery"]
semi_d
```

```python
sns.lineplot(data = semi_d, x= 'Year', y = 'Sales')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data = semi_d, x= 'Year', y = 'Sales')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data = semi_d, x= 'Month', y = 'Sales')
```

```python
semi_d_c = semi_d.copy()
index = []
for days in ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]:
  indx = semi_d[semi_d['Day'] == days].index
  index.append(indx)

index = [v for row in index for v in row]
print(index)
```

```python
semi_d_c.loc[index]
```

```python
semi_d_c = semi_d_c.loc[index]
semi_d_c
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data = semi_d, x= 'Day', y = 'Sales')
```

The days of the week has now been reordered in the plot below

```python
plt.figure(figsize=(20,10))
sns.lineplot(data = semi_d_c, x= 'Day', y = 'Sales')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data = semi_d_c, x= 'Day', y = 'Sales', hue = "Region")
```
