```python
plt.plot(temporary_data['Mes'],temporary_data['data science'], label = 'data science')
plt.plot(temporary_data['Mes'],temporary_data['machine learning'], label = 'machine learning')
plt.plot(temporary_data['Mes'],temporary_data['deep learning'], label = 'deep learning')
plt.xlabel('Date')
plt.ylabel('Popularity')
plt.title('Popularity of AI terms by date')
plt.grid(True)
plt.legend()
```

```python
plt.plot(temporary_data['Mes'],temporary_data['data science'], 'r-') #'r-': This is specifying the style of the line plot. In this case, it's a red line ('r') with a solid line style ('-').
plt.plot(temporary_data['Mes'],temporary_data['data science']*2, 'bs') # 'bs' means you want to use blue squares ('b' for blue and 's' for squares) to represent the data points on the plot.
plt.plot(temporary_data['Mes'],temporary_data['data science']*3, 'g^')
```
